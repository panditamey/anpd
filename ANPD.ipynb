```python
import cv2
from google.colab.patches import cv2_imshow
min_area = 12000
count = 1
max_imgs = 18
img_name = "data2.png"
img = cv2.imread("/content/New folder/Data/"+img_name)
logo = cv2.imread("/content/New folder/logo.png")

haarcascade = "haarcascade.xml"

plate_cascade = cv2.CascadeClassifier(haarcascade)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plates = plate_cascade.detectMultiScale(img_gray, 1.1, 4)

for (x,y,w,h) in plates:
    area = w * h

    if area > min_area:
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
        img_roi = img[y: y+h, x:x+w]
        cv2_imshow(img_roi)

logo = cv2.resize(logo, (w, h))
print(img.shape)
print(logo.shape)
print(w)
print(h)
img[y:y+h, x:x+w] = logo
cv2_imshow(img)
cv2.imwrite("img.jpg",img)
```
